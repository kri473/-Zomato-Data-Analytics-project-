## Exploratory Data Analysis:

***Basic Statistics:***

Calculate and visualize the average rating of restaurants
Analyze the distribution of restaurant ratings to understand the overall rating
landscape

```python
# Average rating of restaurants
avg_rating = data['aggregate_rating'].mean()
print("Average Rating:", avg_rating)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.hist(data["aggregate_rating"], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()

```

## Key Insights
Many restaurants have rating = 0

This is unusual because it’s unlikely that so many restaurants are truly rated “0.”

Most likely, 0 means missing data or not yet reviewed.

Ratings between 2 and 5 are spread more evenly

Customers usually give moderate to high ratings.

Very few restaurants are rated extremely low (like 1).

```python
sns.boxplot(x=data['aggregate_rating'], color='lightgreen')
plt.title("Box Plot of Ratings")
plt.show()
```

👉 Most restaurants are rated between 2 and 4,
👉 The average rating is around 3.5,
👉 The 0 rating is an outlier (likely missing or invalid data).

***Location Analysis:***

Identify the city with the highest concentration of restaurants

```python
c = data['city'].value_counts().head(10)
c
```

```python
c.plot(kind='bar', color='skyblue', figsize=(10,6))
plt.title("Top 10 Cities by Number of Restaurants")
plt.ylabel("Number of Restaurants")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.show()
```

👉 Chennai has the most restaurants, followed by Mumbai and Bangalore.
👉 Goa has the least restaurants among the listed cities.
👉 Overall, big metro cities dominate the restaurant count, showing higher food business activity.

Visualize the distribution of restaurant ratings across different cities

```python
plt.figure(figsize=(12,6))


top_cities = data['city'].value_counts().head(10).index
sns.boxplot(x='city', y='aggregate_rating', data=data[data['city'].isin(top_cities)], palette='Set2')

plt.title("Distribution of Restaurant Ratings Across Top 10 Cities")
plt.xticks(rotation=90)
plt.show()
```

👉 Most cities have restaurant ratings centered around 3.5 (average quality).
👉 The spread is usually between 2 and 4, showing consistency across cities.
👉 The 0 rating appears as an outlier, meaning missing or invalid data.

***Price Range and Rating:***

Analyze the relationship between price range and restaurant ratings.

```python
# Group by price range and calculate average rating
avg_rating_by_price = data.groupby('price_range')['rating'].mean()
print(avg_rating_by_price)
```

```python


# Grouped data
avg_rating_by_price = data.groupby('price_range')['rating'].mean()

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(avg_rating_by_price.index, avg_rating_by_price.values, color='blue', s=100)

# Labels and title
plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.grid(True)
plt.show()
```

👉 As the price range increases, the average rating also goes up.
👉 Customers tend to give higher ratings to more expensive restaurants.
👉 Lower price range restaurants usually get lower average ratings.

***Online Order and Table Booking:***

Investigate the impact of online order availability on restaurant ratings

```python
avg_rating_online = data.groupby('delivery')['rating'].mean()
print(avg_rating_online)
```

```python
plt.figure(figsize=(6,4))
avg_rating_online.plot(kind='bar', color='pink', edgecolor='black')

# Labels and title
plt.title("Average Rating by Delivery Availability")
plt.xlabel("Delivery Available")
plt.ylabel("Average Rating")
plt.xticks(rotation=0)
plt.show()
```

***Top Restaurant Chains***

Identify and visualize the top restaurant chains based on the number of outlets.

```python
# Count outlets per restaurant chain
top_chains = data['name'].value_counts().head(10)
print("Top 10 Restaurant Chains:\n", top_chains)
```

```python


top_chains.plot(kind='bar', color='orange', figsize=(10,6))
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.ylabel("Number of Outlets")
plt.xlabel("Restaurant Chain")
plt.xticks(rotation=45)
plt.show()
```

👉 Domino’s Pizza has the most outlets, showing its strong presence.
👉 Cafe Coffee Day and KFC follow next with large networks.
👉 Barbeque Nation has the fewest outlets among the top 10 chains.

Explore the ratings of these top chains

```python
top_chain_names = top_chains.index
avg_chain_ratings = data[data['name'].isin(top_chain_names)] \
                    .groupby('name')['rating'].mean().sort_values(ascending=False)

print("Average Ratings of Top Chains:\n", avg_chain_ratings)
```

```python
avg_chain_ratings.plot(kind='bar', color='skyblue', figsize=(10,6))
plt.title("Average Ratings of Top Restaurant Chains")
plt.ylabel("Average Rating")
plt.xlabel("Restaurant Chain")
plt.xticks(rotation=45)
plt.show()
```

👉 Barbeque Nation has the highest average rating (above 4.5), showing strong customer satisfaction.
👉 Chains like Burger King, McDonald’s, KFC, Domino’s, Subway, and Pizza Hut have ratings around 3.6–3.9 (good but not excellent).
👉 Baskin Robbins and Café Coffee Day have the lowest ratings (~3.0) among the group.
